```python
BOOK_NAMES = [
    "THE SONNETS",
    "ALL’S WELL THAT ENDS WELL",
    "THE TRAGEDY OF ANTONY AND CLEOPATRA",
    "AS YOU LIKE IT",
    "THE COMEDY OF ERRORS",
    "THE TRAGEDY OF CORIOLANUS",
    "CYMBELINE",
    "THE TRAGEDY OF HAMLET, PRINCE OF DENMARK",
    "THE FIRST PART OF KING HENRY THE FOURTH",
    "THE SECOND PART OF KING HENRY THE FOURTH",
    "THE LIFE OF KING HENRY THE FIFTH",
    "THE FIRST PART OF HENRY THE SIXTH",
    "THE SECOND PART OF KING HENRY THE SIXTH",
    "THE THIRD PART OF KING HENRY THE SIXTH",
    "KING HENRY THE EIGHTH",
    "THE LIFE AND DEATH OF KING JOHN",
    "THE TRAGEDY OF JULIUS CAESAR",
    "THE TRAGEDY OF KING LEAR",
    "LOVE’S LABOUR’S LOST",
    "THE TRAGEDY OF MACBETH",
    "MEASURE FOR MEASURE",
    "THE MERCHANT OF VENICE",
    "THE MERRY WIVES OF WINDSOR",
    "A MIDSUMMER NIGHT’S DREAM",
    "MUCH ADO ABOUT NOTHING",
    "THE TRAGEDY OF OTHELLO, THE MOOR OF VENICE",
    "PERICLES, PRINCE OF TYRE",
    "KING RICHARD THE SECOND",
    "KING RICHARD THE THIRD",
    "THE TRAGEDY OF ROMEO AND JULIET",
    "THE TAMING OF THE SHREW",
    "THE TEMPEST",
    "THE LIFE OF TIMON OF ATHENS",
    "THE TRAGEDY OF TITUS ANDRONICUS",
    "TROILUS AND CRESSIDA",
    "TWELFTH NIGHT; OR, WHAT YOU WILL",
    "THE TWO GENTLEMEN OF VERONA",
    "THE TWO NOBLE KINSMEN",
    "THE WINTER’S TALE",
    "A LOVER’S COMPLAINT",
    "THE PASSIONATE PILGRIM",
    "THE PHOENIX AND THE TURTLE",
]
```

```python
txt = open("shakespeare.txt", "r").read()

books = {}
for book_name, next_book_name in zip(BOOK_NAMES, BOOK_NAMES[1::]):
    book_start = txt.index(book_name)
    book_end = txt.index(next_book_name)
    # print(book_name, book_start, book_end)
    txt = txt[book_end:]
    if "ACT" not in txt[book_start:book_end]:
        pass
    else:
        books[book_name] = txt[book_start:book_end]
```

```python
scenes = []
for book_name, book in books.items():
    for scene in book.split("\n\n\n"):
        if scene in BOOK_NAMES:
            pass
        elif len(scene) < 30:
            pass
        elif "Contents" in scene:
            pass
        else:
            scenes.append(scene.strip("\n"))
```

```python
scene_chunks = []
max_length = 4000
for scene in scenes:
    if len(scene) > max_length:
        for i in range(0, len(scene), max_length):
            scene_chunks.append(scene[i : i + max_length])
    else:
        scene_chunks.append(scene)
```

```python
QUESTION_GENERATION_PROMPT = """
You are reading a chunk of a scene from a shakespeare play.

Please generate 10 questions, separated by newlines, that would be interesting and relevant to ask about the scene.

{scene}
"""
```

```python
print(scene_chunks[15])
```

```python
import ollama
resp = ollama.generate(model="deepseek-r1:8b", system="", prompt=QUESTION_GENERATION_PROMPT.format(scene=scene_chunks[15]))
resp
```

```python
questions = [r.strip() for r in resp.response.split("</think>")[-1].split("\n") if r.strip()]
questions
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("joe32140/ModernBERT-base-msmarco", truncate_dim=512)
content_embeddings = model.encode(scene_chunks, batch_size=1, show_progress_bar=True)
```

```python
content_embeddings
```

```python
scenes[0]
```

```python
query_embeddings = model.encode(questions, batch_size=1, show_progress_bar=True)
```

```python
sim_to_query = model.similarity(content_embeddings, query_embeddings)
import numpy as np

np.argsort(sim_to_query.mean(-1).reshape(-1))[-5::]
```

```python
for i in range(sim_to_query.shape[1]):
    print(questions[i])
    print(np.argsort(sim_to_query[:, i].reshape(-1))[-5::])
    print("\n\n")
```
